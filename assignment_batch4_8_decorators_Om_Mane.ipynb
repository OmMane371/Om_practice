**Q22.** Explain (in a comment) why `convert_temperature()` should be an instance method if it depends on an object’s `scale` (C or F). Optionally, show a small example.

```python
class Temperature:
    def __init__(self, value, scale):
        self.value = value
        self.scale = scale   

    # convert_temperature() should be an instance method because:
    # It depends on the object's own 'scale' value.
    # Each object may have a different scale (one may be Celsius, another Fahrenheit),
    # so the conversion must use the specific object's data.
    # Instance methods have access to 'self', allowing them to read that scale.

    def convert_temperature(self):
        if self.scale == 'C':
            return (self.value * 9/5) + 32  
        elif self.scale == 'F':
            return (self.value - 32) * 5/9  
t1 = Temperature(100, 'C')
t2 = Temperature(212, 'F')

print(t1.convert_temperature())  
print(t2.convert_temperature())  
```

**Q24.** Make a class `Date` with class method `from_string(cls, s)` to parse `"DD-MM-YYYY"` and return an object.

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year

    @classmethod
    def from_string(cls, s):
        d , m, y = s.split("-")
        return cls(int(d), int(m), int(y))



date1 = Date.from_string("25-12-2025")
print(date1.day, date1.month, date1.year)   



```

**Q26.** Show difference between class variable and instance variable with a simple example in one class.

```python
class Example:
    class_var = 100        # class variable (shared by all objects)

    def __init__(self, value):
        self.instance_var = value   # instance variable (different for each object)


# Create two objects
obj1 = Example(10)
obj2 = Example(20)

# Print values
print("obj1:", obj1.class_var, obj1.instance_var)   # 100, 10
print("obj2:", obj2.class_var, obj2.instance_var)   # 100, 20

# Change class variable using class name
Example.class_var = 999

print("After changing class_var:")
print("obj1:", obj1.class_var, obj1.instance_var)   # 999, 10
print("obj2:", obj2.class_var, obj2.instance_var)   # 999, 20



```

## G. Decorators (27–30)


**Q27.** Write a simple decorator that prints `"Function started"` before running the function. Test it on a small function.

```python
def simple_decorater(fun):
    def wrapper (*args,**kwargs):
       print('function started')
       res= fun(*args,**kwargs)
       return res
    return wrapper

@simple_decorater
def add(a,b):
    return a+b
add(1,2)
```

**Q28.** Create a decorator that prints `"Function ended"` after running the function. Test it.

```python
def end(fun):
    def wrapper(*args,**kwargs):
        res=fun(*args,**kwargs)
        print('function ended')
        return res
    return wrapper



@end
def add(a,b):
    return a*b
add(10,20)
```

**Q29.** Make a decorator that prints the **name** of the function being called (use `func.__name__`).

```python
def func_name_(fun):
     def wrapper(*args,**kwargs):
        print('calling function:',fun.__name__)
        return fun (*args,**kwargs)
     return wrapper
        
@func_name_
def add(a,b):
    return a+b

add(20,30)
```

**Q30.** Combine two decorators: one prints `"Start"`, another prints `"End"` around a function. Check the order of messages.

```python

def starting(fun):
  def wrapper(*args,**kwargs):
    print('start')
    return fun(*args,**kwargs)
  return wrapper 

def ending(fun):
  def wrapper(*args,**kwargs):
    pri=fun(*args,**kwargs)
    print('end')
    return pri
  return wrapper 

@starting
@ending
def hello():
  print('Hello')


hello()
```
