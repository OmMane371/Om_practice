## **2. Find the sum of all numbers inside a nested list.**
Example: `[[10,20], [30], [5, 7, 8]]`

````python
lst=[[10,20], [30], [5, 7, 8]] 
sum=0
for i in lst:
    for y in i:
        sum+=y
print(sum)
````

## **3. Count how many even numbers are present in a nested list.**

````python
lst=[[10,20], [30], [5, 7, 8]] 
count=0
for i in lst:
    for y in i:
        if y%2==0:
            count+=1
print(count)

````

## **4. Convert all strings to uppercase inside a nested list.**
Example: `[['hello', 'world'], ['python']]`

````python
lst=[['hello', 'world'], ['python']]
for i in range(len(lst)):
    for y in range(len(lst[i])):
        lst[i][y]=lst[i][y].upper()
        

print(lst)

        
````

## **9. Replace all negative numbers with 0 inside a nested list.**

````python
lst=[[1,-2], [3,4], [-5]] 
for i in range(len(lst)):
    for j in range(len(lst[i])) :
      if lst[i][j]<0:
       lst[i][j]=0
print(lst)     
````
