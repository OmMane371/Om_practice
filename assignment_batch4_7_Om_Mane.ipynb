### Write out_of_stock(items) that returns all products with stock < 20.

```python
products = [
    {"name": "Pen",'price':50000, "stock": 50},
    {"name": "Notebook",'price':55000, "stock": 15},
    {"name": "Pencil",'price':40000, "stock": 8},
    {"name": "Eraser", 'price':60000,"stock": 25}
    {"name": "Pen",'price':40000, "stock": 30}
]


def out_of_stock(items):
    result = []                   
    
    for item in items:
        if item["stock"] < 20:
            result.append(item)
    
    return result

out_of_stock(products)
```

### Write total_inventory_value(items) → sum(price * stock).

```python
def total_inventory_value(items):
    total=0
    for i in items:
         total += i["price"] * i["stock"]
    return total

total_inventory_value(products)
```
